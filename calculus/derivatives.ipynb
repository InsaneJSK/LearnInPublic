## Derivatives

```python
import sympy as sp
x, y = sp.symbols('x y')
eq = x**3
derivative = sp.diff(eq, x)
print(f"Derivative of {eq} with respect to x is: {derivative}")
```

```python
answer = derivative.subs(x, 2)
print(f"The slope of the tangent line at x = 2 is: {answer}")
```

### Power Rule

```python
x, y = sp.symbols("x y")
eq = x**3
eq
```

```python
eq.diff(x)
```

### Chain Rule

```python
eq = (x**2 + 8)**4
eq
```

```python
eq.diff(x)
```

```python
eq = sp.sqrt(x**2 + 8)
eq
```

```python
eq.diff(x)
```

```python
eq = sp.sin(x**3)
eq
```

```python
eq.diff(x)
```

### Product Rule

```python
eq = (x**2 + 8)*(x**3 + 5)
eq
```

```python
eq.diff(x)
```

### Quotient Rule

```python
eq = (x**3 - 1)/(x**2 - 1)
eq
```

```python
eq.diff(x)
```

```python
eq.diff(x).subs(x, 2)
```
