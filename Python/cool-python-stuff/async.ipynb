## Async Programming

```python
from multiprocessing import Process
```

```python
def myfunc():
    pass
```

```python
procs = []
for _ in range(5):
    procs.append(Process(target=myfunc, args=()))
```

```python
#For starting those functions
for proc in procs:
    proc.start()
```

```python
for proc in procs:
    proc.join()
```

## Multi Threading

```python
from threading import Thread
```

```python
f1 = Thread(target=myfunc)
```

```python
f1.start() #Async method
f1.join() #Sync method
```

## Coroutines

```python
def sample(name):
    try:
        while True:
            what = (yield)
            print(f"{what} {name}")
    except GeneratorExit:
        print("Ok!")
```

## AsyncIO

```python
import asyncio
```

```python
async def main():
    print("Hello")
    await asyncio.sleep(1)
    print("World")
```

```python
type(main())
```

```python
if __name__ == "__main__":
    try:
        asyncio.run(main())  #The line that should work in a py-kernel but due to how main loop works in ipy, it doesn't work in jupyter
    except RuntimeError:
        await main()
```

```python
async def waiter(n):
    await asyncio.sleep(n)
    print(f"Waited for {n} sec")
```

```python
import time
async def main():
    print(time.strftime('%X'))
    await waiter(2)
    await waiter(3)
    print(time.strftime('%X'))
```

```python
if __name__ == "__main__":
    await main()
```

```python
async def main():
    task1 = asyncio.create_task(waiter(2))
    task2 = asyncio.create_task(waiter(3))

    print(time.strftime('%X'))
    await task1
    await task2
    print(time.strftime('%X'))
```

```python
if __name__ == "__main__":
    await main()
```

```python
import aiohttp
```

```python
async def fetchfromgoogle():
    url = 'https://www.google.com'
    session = aiohttp.ClientSession()
    resp = await session.get(url)
    print(await resp.content.read())
    await session.close()
```

```python
async def main():
    for _ in range(20):
        await fetchfromgoogle()
```

```python
if __name__ == '__main__':
    await main()
```

```python
async def main():
    await asyncio.gather(*[fetchfromgoogle() for _ in range(20)])
```

```python
if __name__ == '__main__':
    await main()
```
