## MNIST Dataset

```python
df = pd.read_csv("mnist_train.csv")
```

```python
df.head()
```

```python
data = df.values
```

```python
X = df.iloc[:, 1:].values
y = df.iloc[:, 0].values
```

```python
print(y)
```

```python
def imageiter(end, start = 0):
    while start < end:
        yield start
        start += 1
```

```python
imagegen = imageiter((X.shape[0]))
```

```python

import matplotlib.pyplot as plt
```

```python
i = next(imagegen)
plt.imshow(X[i, :].reshape(28, 28), cmap = "grey")
plt.title(f"Label {y[i]}")
plt.show()
```

```python
plt.figure(figsize=(10, 10))
for k in range(1, 626):
    plt.subplot(25, 25, k)
    i = next(imagegen)
    plt.imshow(X[i, :].reshape(28, 28), cmap = "grey")
    plt.axis("off")
plt.show()
```

```python
## Movie Dataset
```

```python
df = pd.read_csv("movie_metadata.csv")
```

```python
list(df.columns)
```

```python
df.describe()
```

```python
titles = list(df.get("movie_title"))
```

```python
dicti = {}
for t in titles:
    if len(t) in dicti.keys():
        dicti[len(t)] += 1
    else:
        dicti[len(t)] = 1
```

```python
print(dicti)
```

```python
plt.scatter(np.array(list(dicti.keys())), np.array(list(dicti.values())))
plt.title("Movie Name length freq")
plt.xlabel("Number of characters")
plt.ylabel("Number of movies")
plt.show()
```
